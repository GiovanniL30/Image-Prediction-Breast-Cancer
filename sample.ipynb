```python
import matplotlib.pyplot as plt
import random
from PIL import Image
import os

def show_sample_images(base_dir, class_labels=["0", "1"], num_samples=3):
    plt.figure(figsize=(8, 5))
    for class_idx, label in enumerate(class_labels):
        class_path = os.path.join(base_dir, label)
        images = os.listdir(class_path)
        samples = random.sample(images, min(num_samples, len(images)))
        
        for i, img_file in enumerate(samples):
            img_path = os.path.join(class_path, img_file)
            img = Image.open(img_path).resize((224, 224))
            
            plt.subplot(len(class_labels), num_samples, class_idx * num_samples + i + 1)
            plt.imshow(img)
            plt.axis('off')
            plt.title(f"Class {label}")
    
    plt.tight_layout()
    plt.show()

# Show samples from training set
print("Training Set Samples:")
show_sample_images("dataset/train")

# Show samples from validation set
print("Validation Set Samples:")
show_sample_images("dataset/validation")

# Show samples from test set
print("Test Set Samples:")
show_sample_images("dataset/test")
```
